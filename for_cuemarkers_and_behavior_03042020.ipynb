```python
#behavior!
data_read=pd.read_csv('/data/MoodGroup/07M0021_meg_analysis/MID_data/subjects/sub-1051/behavior/MID1-1051-1_behavior.txt',delimiter='	',skiprows=1,encoding='utf-16',engine='python')
```

```python
behavior_1051=data_read[['Cue','ResponseType','TrialCategory','Target.RT']]
behavior_1051
```

```python
behavior_1051_win=behavior_1051[behavior_1051.Cue=='Win2']
behavior_1051_win_hit=behavior_1051_win[behavior_1051_win.ResponseType=='GoodResponse']
behavior_1051_win_hit
```

```python
behavior_1051_win_miss=behavior_1051_win[behavior_1051_win.ResponseType=='NoResponse']
behavior_1051_win_miss
```

```python
behavior_1051_lose=behavior_1051[behavior_1051.Cue=='Lose2']
behavior_1051_lose_hit=behavior_1051_lose[behavior_1051_lose.ResponseType=='GoodResponse']
behavior_1051_lose_hit
```

```python
behavior_1051_lose_miss=behavior_1051_lose[behavior_1051_lose.ResponseType=='NoResponse']
behavior_1051_lose_miss
```

```python
behavior_1051_cont=behavior_1051[behavior_1051.Cue=='Control']
behavior_1051_cont_hit=behavior_1051_cont[behavior_1051_cont.ResponseType=='GoodResponse']
behavior_1051_cont_hit
```

```python
behavior_1051_cont_miss=behavior_1051_cont[behavior_1051_cont.ResponseType=='NoResponse']
behavior_1051_cont_miss
```

```python
win_hit_RT_1051=behavior_1051_win_hit['Target.RT'][behavior_1051_win_hit['Target.RT']!=0].mean()
win_hit_acc_1051=behavior_1051_win_hit.count()
win_miss_RT_1051=behavior_1051_win_miss['Target.RT'][behavior_1051_win_miss['Target.RT']!=0].mean()
win_miss_acc_1051=behavior_1051_win_miss.count()
lose_hit_RT_1051=behavior_1051_lose_hit['Target.RT'][behavior_1051_lose_hit['Target.RT']!=0].mean()
lose_hit_acc_1051=behavior_1051_lose_hit.count()
lose_miss_RT_1051=behavior_1051_lose_miss['Target.RT'][behavior_1051_lose_miss['Target.RT']!=0].mean()
lose_miss_acc_1051=behavior_1051_lose_miss.count()
cont_hit_RT_1051=behavior_1051_cont_hit['Target.RT'][behavior_1051_cont_hit['Target.RT']!=0].mean()
cont_hit_acc_1051=behavior_1051_cont_hit.count()
cont_miss_RT_1051=behavior_1051_cont_miss['Target.RT'][behavior_1051_cont_miss['Target.RT']!=0].mean()
cont_miss_acc_1051=behavior_1051_cont_miss.count()
```

```python
d_1051={'Type':['Win','Win','Lose','Lose','Cont','Cont'],
   'Outcome':['Hit','Miss','Hit','Miss','Hit','Miss'],
   'RT':[win_hit_RT_1051,win_miss_RT_1051,lose_hit_RT_1051,lose_miss_RT_1051,cont_hit_RT_1051,cont_miss_RT_1051],
   'Acc':[win_hit_acc_1051.Cue,win_miss_acc_1051.Cue,lose_hit_acc_1051.Cue,lose_miss_acc_1051.Cue,cont_hit_acc_1051.Cue,cont_miss_acc_1051.Cue]}
df_1051=pd.DataFrame(d_1051)
df_1051=df_1051[['Type','Outcome','RT','Acc']]
df_1051
```

```python
df_1051.to_csv('/data/MoodGroup/07M0021_meg_analysis/MID_data/subjects/sub-1051/behavior/behavior_1051_MID.csv',index=False)
```
